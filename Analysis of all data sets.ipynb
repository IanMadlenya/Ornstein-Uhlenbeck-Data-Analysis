```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
filelist=[file for file in os.listdir(".") if file.endswith('.csv')]
```

```python
result_array = None
```

```python
for file in filelist:
    firstUS = file.find('_')+1
    secondUS = file.find('_',firstUS)
    firstPeriod = file.find('.',secondUS)
    deltaT = float(file[firstUS:secondUS])
    N = int(file[secondUS+1:firstPeriod])
    
    BM = pd.read_csv(file)
    # clean the data to not include fits that unrealistic
    BM = BM[np.abs(BM['decay'])<1000]

    results = [deltaT, N,
               BM['A50'].mean(), BM['A50'].std(), BM['Astd'].mean(),
               BM['amplitude'].mean(), BM['amplitude'].std(), BM['amplitude_std'].mean(),
               BM['D50'].mean(), BM['D50'].std(), BM['Dstd'].mean(),
               BM['decay'].mean(), BM['decay'].std(), BM['decay_std'].mean()]
    print(deltaT, N, results)
    if result_array is None:
        result_array = results
    else:
        result_array = np.vstack((result_array, results))

columns = ['deltaT',
           'N',
           'A50mean',
           'A50std',
           'AstdMean',
           'fitAmpMean',
           'fitAmpStd',
           'fitAmpStdMean',
           'D50mean',
           'D50std',
           'DstdMean',
           'fitDecayMean',
           'fitDecayStd',
           'fitDecayStdMean'
           ]

df = pd.DataFrame(data = result_array, columns=columns)
df.to_csv('AllData.csv', index=False)
```
