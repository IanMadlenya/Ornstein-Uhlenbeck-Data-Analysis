```python
N=10000
#calculate B_max
coef = [(N-1)*BM10000clean['datasq'][0],
       (2.0-N)*BM10000clean['datacorr'][0],
       -BM10000clean['data1sq'][0]-(N+1)*BM10000clean['datasq'][0]-BM10000clean['dataNsq'][0]**2,
       N*BM10000clean['datacorr'][0]]
f = np.poly1d(coef)
```

```python
print(np.roots(coef))
```

```python
B_list=[]
sigsq_list=[]
for i in range(BM10000clean.shape[0]):
    coef = [(N-1)*BM10000clean['datasq'][i],
       (2.0-N)*BM10000clean['datacorr'][i],
       -BM10000clean['data1sq'][i]-(N+1)*BM10000clean['datasq'][i]-BM10000clean['dataNsq'][i]**2,
       N*BM10000clean['datacorr'][i]]
    B=np.roots(coef)[-1]
    B_list.append(B)
    Q=(BM10000clean['data1sq'][i]+BM10000clean['dataNsq'][i])/(1-B**2)
    Q=Q+BM10000clean['datasq'][i]*(1+B**2)/(1-B**2)
    Q=Q-BM10000clean['datacorr'][i]*2*B/(1-B**2)
    sigsq_list.append(Q/N)
B_list=np.array(B_list)
sigsq_list=np.array(sigsq_list)
```

```python
import matplotlib 
matplotlib.rcParams.update({'font.size': 16})
matplotlib.rc('xtick', labelsize=16) 
matplotlib.rc('ytick', labelsize=16) 
plt.hist(B_list,30,normed=True)
plt.xlabel('B')
plt.ylabel('p(B)')
```
