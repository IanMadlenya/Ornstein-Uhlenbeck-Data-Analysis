```python
N=5000
B_list=[]
sigsq_list=[]
for (data1sq,dataNsq,datasq,datacorr) in zip(BMclean['data1sq'],
                                             BMclean['dataNsq'],
                                             BMclean['datasq'],
                                             BMclean['datacorr']):
    coef = [(N-1)*datasq,
       (2.0-N)*datacorr,
       -data1sq-(N+1)*datasq-dataNsq,
       N*datacorr]
    B=np.roots(coef)[-1]
    B_list.append(B)
    Q=(data1sq+dataNsq)/(1-B**2)
    Q=Q+datasq*(1+B**2)/(1-B**2)
    Q=Q-datacorr*2*B/(1-B**2)
    sigsq_list.append(Q/N)
B_list=np.array(B_list)
sigsq_list=np.array(sigsq_list)
```

```python
import matplotlib 
matplotlib.rcParams.update({'font.size': 16})
plt.hist(B_list,30,normed=True)
plt.xlabel('B')
plt.ylabel('p(B)')
```

```python
print(np.exp(-0.01))
print(B_list.mean())
```

```python
import matplotlib 
matplotlib.rcParams.update({'font.size': 16})
plt.hist(BMclean['amplitude'],30,normed=True,label='LS fit')
plt.hist(sigsq_list,30,normed=True,label='Bayesian')
plt.xlabel(r'$\sigma^{2}$')
plt.ylabel(r'$p( \sigma^{2})$')
plt.legend()
```

```python
print(sigsq_list.mean(),sigsq_list.std())
print (BMclean['amplitude'].mean(),BMclean['amplitude'].std())
```

```python
tau_list=-0.01/np.log(B_list)
```

```python
plt.hist(BMclean['decay'],30,normed=True,label='LS fit')
plt.hist(tau_list,30, normed=True,label='Bayesian')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$p(\tau)$')
plt.legend
```

```python
ff=[0,4]
plt.plot(np.array(BMclean['decay']),tau_list,"o")
plt.plot(ff,ff)
plt.xlabel(r'$\tau$ least square fit')
plt.ylabel(r'$\tau$ Bayesian')
```
