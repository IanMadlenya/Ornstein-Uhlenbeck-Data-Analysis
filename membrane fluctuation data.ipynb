```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dACF_101216ves04 = np.genfromtxt("DODS_data/dACF_101216ves04.txt", skip_header=1)
dhvst_101216ves04 = np.genfromtxt("DODS_data/dhvst_101216ves04.txt", skip_header=1)
```

```python
plt.plot(dhvst_101216ves04.T[0],dhvst_101216ves04.T[1])
print("standard deviation:", dhvst_101216ves04.T[1].std())
print("variance:", dhvst_101216ves04.T[1].std()**2)
```

```python
plt.semilogx(dACF_101216ves04.T[0],dACF_101216ves04.T[1])
```
